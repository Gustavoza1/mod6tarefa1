```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#%matplotlib notebook
#%matplotlib inline
```

```python
base = pd.read_csv("b1abb33c4f.csv")
base
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
#Fiquei com duvida sobre a descricao da pergunta, ˜monte *um* graficoS com as variaveiS˜, devido a isso fiz a criacao dos graficoos de forma separada.

#APGR1
plt.figure(figsize=(8, 6))
sns.countplot(x='APGAR1', data=base)
plt.title('APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Contagem')
plt.show()

#APGR5
plt.figure(figsize=(8, 6))
sns.countplot(x='APGAR5', data=base)
plt.title('APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Contagem')
plt.show()

#SEXO
plt.figure(figsize=(8, 6))
sns.countplot(x='SEXO', data=base)
plt.title('SEXO')
plt.xlabel('SEXO')
plt.ylabel('Contagem')
plt.show()

#GESTACAO
plt.figure(figsize=(8, 6))
sns.countplot(x='GESTACAO', data=base)
plt.xticks(rotation=30) #nao funcionou a rotacao dos valores
plt.title('GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Contagem')
plt.show()

#ESTCIVMAE
plt.figure(figsize=(8, 6))
sns.countplot(x='ESTCIVMAE', data=base)
plt.title('ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Contagem')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=base, estimator='mean')
plt.title('APGAR5 Médio por Consultas')
plt.xlabel('Consultas')
plt.ylabel('APGAR5 Médio')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=base)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=base)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=base)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=base)
plt.tight_layout() #ajuste sobreposicao
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=30)  # Adicionando rotação de 30 graus
#ao utilizar o codigo acima para rotacionar o eixo x do grafico GESTACAO, e adicionado um texto com os valores antes dos graficos.. nao melhorando a visualizacao.
axes[0, 1].set_xlabel('GESTACAO', fontsize=10) #tentativa de reduzir eixo do grafico GESTACAO

plt.subplots_adjust(wspace=0.1, hspace=0.5) #ajuste no espaco entre os graficos para melhorar a visualizao do rotulo do eixo GESTACAO
```
